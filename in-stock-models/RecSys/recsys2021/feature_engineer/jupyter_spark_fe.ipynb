```python
import init

import findspark
findspark.init()

import os
import pandas as pd
import numpy as np
from pyspark.sql import *
from pyspark import *
import pyspark.sql.functions as f
from timeit import default_timer as timer
from pyrecdp.data_processor import *
from pyrecdp.utils import *

spark = SparkSession.builder.master('yarn')\
        .appName("Recsys2021_data_process")\
        .getOrCreate()
```

```python
current_path = "/recsys2021_0608_fe_processed/"
path_prefix = "hdfs://"

data_path= "/recsys2021_0608_fe_sample0.1/dask_fe_processed/train_fe_processed/"
tgt_train_df = spark.read.parquet("%s/" % (data_path))
#tgt_train_df.show(1, vertical=True, truncate=False)

data_path= "/recsys2021_0608_fe_sample0.1/dask_fe_processed/valid_fe_processed/"
tgt_valid_df = spark.read.parquet("%s/" % (data_path))
#tgt_valid_df.show(1, vertical=True, truncate=False)
```

```python
train_spark_path= "/recsys2021_0608_fe_processed/train_fe/"
train_df = spark.read.parquet("%s/" % (train_spark_path))
#train_df.show(1, vertical=True, truncate=False)
```

```python
valid_spark_path= "/recsys2021_0608_fe_processed/valid_fe/"
valid_df = spark.read.parquet("%s/" % (valid_spark_path))
#valid_df.show(1, vertical=True, truncate=False)
```

```python
fe_cols = [
    'tweet_id'                                               ,
    'enaging_user_id'                                        ,
    'CE_enaging_user_id_norm'                                ,
    #'TE_tweet_type_reply_timestamp'                          ,                                                              
    #'TE_language_reply_timestamp'                            ,                                                              
    #'TE_engaged_with_user_id_reply_timestamp'                ,                                                              
    #'TE_enaging_user_id_reply_timestamp'                     ,                                                              
    #'TE_present_media_retweet_timestamp'                     ,                                                              
    #'TE_tweet_type_retweet_timestamp'                        ,                                                              
    #'TE_language_retweet_timestamp'                          ,                                                              
    #'TE_engaged_with_user_id_retweet_timestamp'              ,                                                              
    #'TE_enaging_user_id_retweet_timestamp'                   ,                                                              
    #'TE_present_media_retweet_with_comment_timestamp'        ,                                                              
    #'TE_tweet_type_retweet_with_comment_timestamp'           ,                                                              
    #'TE_language_retweet_with_comment_timestamp'             ,                                                              
    #'TE_engaged_with_user_id_retweet_with_comment_timestamp' ,                                                              
    #'TE_enaging_user_id_retweet_with_comment_timestamp'      ,                                                              
    #'TE_present_media_like_timestamp'                        ,                                                              
    #'TE_tweet_type_like_timestamp'                           ,                                                              
    #'TE_language_like_timestamp'                             ,                                                              
    #'TE_engaged_with_user_id_like_timestamp'                 ,                                                              
    #'TE_enaging_user_id_like_timestamp'                                            
]
```

```python
tgt_tmp_df = tgt_train_df.select(*fe_cols)
train_tmp_df = train_df.select(*fe_cols)

res = train_tmp_df.join(tgt_tmp_df, ['tweet_id', 'enaging_user_id'], 'left')
res.show(vertical=True)
```

```python
tgt_tmp_df = tgt_valid_df.select(*fe_cols)
valid_tmp_df = valid_df.select(*fe_cols)

res = valid_tmp_df.join(tgt_tmp_df, ['tweet_id', 'enaging_user_id'], 'left')
res.show(vertical=True)
```

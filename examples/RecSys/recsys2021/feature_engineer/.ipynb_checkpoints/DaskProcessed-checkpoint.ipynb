```python
for name, feature in feature_map.items():
    start = time.time()
    join_cols = eval(name)       
    expr_list = [F.last(F.col(c)).alias(c) for c in feature]
    df = train.select(['fold'] + join_cols + feature).groupby(['fold'] + join_cols).agg(*expr_list)
    out_name = "lookup_table_%s" % "_".join(join_cols)
    path = "file:///mnt/nvme2/chendi/BlueWhale/DaskProcessed0_0_3/train_lookup/"
    df.write.format('parquet').mode('overwrite').save("%s/%s" % (path, out_name))
    print("%s took %.3f secs" % (out_name, (time.time() - start)))
    
```

```python
import hashlib
def clean_text(text):
        if len(text) > 1:
            if text[-1] in ['!', '?', ':', ';', '.', ',']:
                return(text[:-1])
        return(text)

def extract_hash(text, split_text='@', no=0):
        text = text.lower()
        uhash = ''
        text_split = text.split('@')
        if len(text_split) > (no+1):
            text_split = text_split[no+1].split(' ')
            cl_loop = True
            uhash += clean_text(text_split[0])
            while cl_loop:
                if len(text_split) > 1:
                    if text_split[1] in ['_']:
                        uhash += clean_text(text_split[1]) + \
                            clean_text(text_split[2])
                        text_split = text_split[2:]
                    else:
                        cl_loop = False
                else:
                    cl_loop = False
        hash_object = hashlib.md5(uhash.encode('utf-8'))
        return hash_object.hexdigest()
to_notsign = F.udf(lambda x: x.replace('\[CLS\] RT @', ''))
user_defined_hash = F.udf(
        lambda x: extract_hash(x, split_text='RT @', no=0))
user_define_hash_1 = F.udf(lambda x: extract_hash(x))
user_define_hash_2 = F.udf(lambda x: extract_hash(x, no=1))


valid = spark.read.parquet("/recsys2021_0608_processed/validate_decoded_with_word")
valid = valid.withColumn('dt_dow', F.dayofweek(F.from_unixtime(F.col('tweet_timestamp'))).cast(IntegerType()))\
             .withColumn('dt_hour', F.hour(F.from_unixtime(F.col('tweet_timestamp'))).cast(IntegerType()))\
             .withColumn('dt_minute', F.minute(F.from_unixtime(F.col('tweet_timestamp'))).cast(IntegerType()))\
             .withColumn('dt_second', F.second(F.from_unixtime(F.col('tweet_timestamp'))).cast(IntegerType()))\
             .withColumn('a_ff_rate', F.col("engaged_with_user_following_count")/F.col("engaged_with_user_follower_count"))\
             .withColumn('b_ff_rate', F.col("engaging_user_following_count") /F.col("engaging_user_follower_count"))\
             .withColumn('has_rt', F.col('tweet').rlike("^RT\s+|\sRT\s+"))
valid = valid.fillna(0, ['reply_timestamp', 'retweet_timestamp', 'retweet_with_comment_timestamp', 'like_timestamp'])
valid = valid.withColumn('reply_timestamp', F.when(F.col('reply_timestamp') > 0, 1).otherwise(0))\
             .withColumn('retweet_timestamp', F.when(F.col('retweet_timestamp') > 0, 1).otherwise(0))\
             .withColumn('retweet_with_comment_timestamp', F.when(F.col('retweet_with_comment_timestamp') > 0, 1).otherwise(0))\
             .withColumn('like_timestamp', F.when(F.col('like_timestamp') > 0, 1).otherwise(0))
valid = valid.fillna("", ['tweet'])
valid = valid.withColumn("tweet_nortsign", to_notsign(F.col('tweet')))\
             .withColumn('tw_uhash', user_defined_hash(F.col('tweet')))\
             .withColumn('tw_hash', F.hash(F.col('tweet'))%1000000000)\
             .withColumn('tw_hash0', user_define_hash_1(F.col('tweet_nortsign')))\
             .withColumn('tw_hash1', user_define_hash_2(F.col('tweet_nortsign')))
tw_hash_df = spark.read.parquet("/recsys2021_0608_processed/recsys_dicts/tw_hash")
tw_hash_df = tw_hash_df.withColumnRenamed("dict_col", "tw_hash").withColumnRenamed("dict_col_id", "categorified_tw_hash")
valid = valid.join(tw_hash_df, 'tw_hash', 'left').drop('tw_hash').withColumnRenamed('categorified_tw_hash', 'tw_hash')

def merge_indicator_by_input_cols(proc, df, te, is_train=True, out_name="validate_with_features"):
    start = time.time()
    op_merge = ModelMerge(te)
    proc.reset_ops([op_merge])
    proc.transform(df, name=out_name)
    if is_train:
        print("Train Merge models took: %.3f secs" % (time.time()-start))
    else:
        print("Valid Merge models took: %.3f secs" % (time.time()-start))


input_cols_1 = [
    (['present_media'], 'lookup_table_present_media'), 
    (['tweet_type'], 'lookup_table_tweet_type'), 
    (['language'], 'lookup_table_language'), 
    (['engaged_with_user_id'], 'lookup_table_engaged_with_user_id'), 
    (['engaging_user_id'], 'lookup_table_engaging_user_id'), 
    (['present_domains', 'language', 'engagee_follows_engager', 'tweet_type', 'present_media', 'engaged_with_user_is_verified'],
     'lookup_table_present_domains_language_engagee_follows_engager_tweet_type_present_media_engaged_with_user_is_verified'),
    (['engaged_with_user_id', 'tweet_type', 'language'], 
     'lookup_table_engaged_with_user_id_tweet_type_language'),
    (['tw_first_word', 'tweet_type', 'language'], 'lookup_table_tw_first_word_tweet_type_language'), 
    (['tw_last_word', 'tweet_type', 'language'], 'lookup_table_tw_last_word_tweet_type_language'), 
    (['tw_hash0', 'tweet_type', 'language'], 'lookup_table_tw_hash0_tweet_type_language'), 
    (['tw_hash1', 'tweet_type', 'language'], 'lookup_table_tw_hash1_tweet_type_language'), 
    (['tw_uhash', 'tweet_type', 'language'], 'lookup_table_tw_uhash_tweet_type_language'), 
    (['tw_hash'], 'lookup_table_tw_hash'), 
    (['present_media', 'tweet_type', 'language', 'engaged_with_user_is_verified', 'engaging_user_is_verified', 'engagee_follows_engager'], 
     'lookup_table_present_media_tweet_type_language_engaged_with_user_is_verified_engaging_user_is_verified_engagee_follows_engager'), 
    (['present_domains', 'present_media', 'tweet_type', 'language'], 
     'lookup_table_present_domains_present_media_tweet_type_language'), 
    (['present_links', 'present_media', 'tweet_type', 'language'], 
     'lookup_table_present_links_present_media_tweet_type_language'), 
    (['hashtags', 'present_media', 'tweet_type', 'language'], 
     'lookup_table_hashtags_present_media_tweet_type_language')
]


############## Train join #############
current_path = "/mnt/nvme2/chendi/BlueWhale/DaskProcessed0_0_3/"
path_prefix="file://"
valid_out_dfs = []
proc = DataProcessor(spark, path_prefix,
                     current_path=current_path, dicts_path=dicts_folder)

valid_res_dfs = {}
for in_cols, dict_name in input_cols_1:
    valid_res_dfs[dict_name] = (in_cols, spark.read.parquet("%s/%s/valid_lookup/%s" % (path_prefix, current_path, dict_name)), 0)
merge_indicator_by_input_cols(proc, valid, valid_res_dfs, is_train=False, out_name = "validate_with_features")
```
